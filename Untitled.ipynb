There are 99 experiments, each with some number of measurements.  29 is not the maximum number of measurements per experiment (or overall).  It's only the highest "measurement number", whatever that means.  Each experiment contains numerous cycles of measurement where measurement numbers 1 up to 29 are recorded in each cycle.  Cycles are not of equal length though, as we see below, the number of cycles per experiment is fairly constant.

```python
for en in range(0, 98):
    # counting the number of '0' labled measurements gives the number of measurement cycles
    # because they always start at 0 and progress from there
    d = data[['measurement_number', 'experiment_number', 'robustness']]\
        [(data.measurement_number == 0) & (data.experiment_number == en)]
    data.set_value(data.loc[data.experiment_number == en].index, 'cycles', len(d))
    
fig, ax = plt.subplots(1,1, figsize=(15,5))
data[['cycles', 'experiment_number']][data.cycles.notnull()].drop_duplicates()\
    .sort_values('experiment_number')\
    .plot(kind='bar', x='experiment_number', y='cycles', ax=ax);
ax.set_title('number of measurement cycles for each experiment');
```

Measurement cycles are also of different lengths.  That is, there can be as few as 4 measurements per cycle, or as many as 29.  I don't know if there's a reason to care about that.

Positions: 
- ``Joint1 (all fingers) (red)    tends to inhabit the leftward position (x axis here seems to be a range of motion)``
- ``Joint2 (all fingers) (orange) tends to inhabit the center positions``
- ``Joint3 (all fingers) (blue)   tends to inhabit the rightward positions``

Velocities:
- ``Joint1 (all fingers) (red)    tends towards 0 velocity``
- ``Joint2 (all fingers) (orange) tends towards a narrow range of velocities around 0, +- 10``
- ``Joint3 (all fingers) (blue)   tends towards a wider range of velocities around 0, +- 30``

Effort:
- ``Joint1 (all fingers) (red)    tends towards little effort (consistent with its low velocity)``
- ``Joint2 (all fingers) (orange) tends towards high effort and has the widest range of effort``
- ``Joint3 (all fingers) (blue)   tends towards middle effort``

The sense overall is that:
* Joint3 operates at the outside of the range of motion, producing higher velocities but making modest effort.  This is where the grasping is probably at it's most sensative.
* Joint2 operates in a narrower range of motion, producing lower velocities but requiring the most effort.  This is where the heavy lifting is done.
* Joint1 operates in the narrowest range of motion, barely moving and requiring almost no effort.  This must be close to the base, doing little more than determining the angle of the finger.


* Robustness seems to have threasholds between .2 to .4 and between .8 to .99. 
* High robustness does not seem associated with these positions and ranges as such, as they are roughly the same in both robust and non-robust circumstances
* High robustness does seem associated with a 'cleaner' distribution of these values. That is, when robustness is high, the colors bleed into each other less.


* Why are there velocities at all, if the object is always the same?  Because the hand is being shaken, forcing the ball to push one finger then the other.  
* Are finger velocities absolute (relative to the table) or relative to each other?


* Presumably, the goal is to adjust low and medium robustness situations to high robustness, so the point is not just to identify what robust looks like but to chart a course from one state to the other
* Since position is the most goal-like value here, a strategy might be, for any given position, to find the next one that is most similar in shape but also more robust, altering effort as needed to achieve that new position. Then move up the steps until high robustness is reached.  Judge effort by looking at the mean of the difference in effort between the samples similar to the current position and samples similar to the one that is sought, then adding that difference to the current effort.  Presumably, velocity is not under our direct control.
